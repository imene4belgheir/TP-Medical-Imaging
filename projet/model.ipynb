### CELL 2: Define Dataset Path & Load Images

```python
# Define the main dataset folder path
DATASET_PATH = Path('dataset')  

# Define class names (folder names must match exactly)
CLASS_NAMES = ['glioma_tumor', 'meningioma_tumor', 'pituitary_tumor', 'no_tumor_cleaned_']
CLASS_LABELS = {name: idx for idx, name in enumerate(CLASS_NAMES)}

print("Class Mapping:")
for name, idx in CLASS_LABELS.items():
    print(f"  {idx}: {name.capitalize()}")

# Verify dataset structure
print("\n📁 Dataset Structure Verification:")
for class_name in CLASS_NAMES:
    class_path = DATASET_PATH / class_name
    if class_path.exists():
        num_images = len(list(class_path.glob('*.jpg'))) + len(list(class_path.glob('*.JPG')))
        print(f"  ✓ {class_name}/: {num_images} images found")
    else:
        print(f"  ✗ {class_name}/: FOLDER NOT FOUND")
```
